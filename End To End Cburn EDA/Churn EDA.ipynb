```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Customer Churn.csv')
data
```

```python
# Replace the white space in total charges columns with 0 
data['TotalCharges'] = data['TotalCharges'].replace(" " , '0')
```

```python
data[data['tenure']==0]
```

```python
# convert total charges column into float

data['TotalCharges'] = data['TotalCharges'].astype('float')
```

```python
# duplicated Values in data
data.duplicated().sum()
```

```python
# modify columns Senior Citizen
data['SeniorCitizen'].value_counts()
```

```python
# Replace 1 with yes and else with 0

def con(value):
    if value == 1:
        return "yes"
    else:
        return "no"
```

```python
data['SeniorCitizen'] = data['SeniorCitizen'].apply(con)
```

```python
data['SeniorCitizen'].value_counts()
```
