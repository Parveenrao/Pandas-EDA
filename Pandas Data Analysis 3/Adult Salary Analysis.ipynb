## 7. Perform Data Cleaning(Replace "?" with NaN)

```python
data.tail(20)
```

```python
data.isin(['?']).sum()
```

```python
import numpy as np
data['workclass'] = data['workclass'].replace('?' , np.nan)
data['occupation'] = data['occupation'].replace('?', np.nan)
data['native-country'] = data['native-country'].replace('?' , np.nan)
```

```python
data.tail(20)
```

```python
data.isin(['?']).sum()
```
